## Step 2: Import required libraries

```python
import pandas as pd
from pycaret.classification import *

import warnings
warnings.filterwarnings('ignore')
```

## Step 3: Import the train set of the titanic dataset




```python
# Load the Drive helper and mount
from google.colab import drive
drive.mount('/content/drive')
```

```python
train_df = pd.read_csv('/content/drive/My Drive/Applied Machine Learning/Datasets/titanic_train.csv')
train_df.head(10)
```

## Step 4: Setup the PyCaret environment

```python
clf = setup(data= train_df, target= 'Survived', session_id= 123)
```

## Step 5: Compare all models



```python
models = compare_models()
```

## step 6:  Select the best model

```python
best_model = automl(optimize = 'Accuracy')
best_model
```

## Step 7: Fine-tuning the best model


```python
tuned_best_model = tune_model(best_model)
tuned_best_model
```

## Step 8: Evaluate the best model

```python
evaluate_model(tuned_best_model)
```

## Step 9: Save the tuned best model

```python
save_model(tuned_best_model, '/content/drive/My Drive/Applied Machine Learning/Chapter7/Streamlit Project/pycaret_titanic_tuned_best_model')
```

## Step 10: Load the tuned best model

```python
imported_model = load_model('/content/drive/My Drive/Applied Machine Learning/Chapter7/Streamlit Project/pycaret_titanic_tuned_best_model')
```

## Step 11: Import the test set of titanic dataset

```python
test_df = pd.read_csv('/content/drive/My Drive/Applied Machine Learning/Datasets/titanic_test.csv')
test_df.head(10)
```

## Step 12: Make predictions and Evaluate the model

```python
prediction = predict_model(imported_model, data = test_df)
prediction.rename(columns={'prediction_label': 'Predicted_Survived', 'prediction_score': 'Predicted_Survived_Score'}, inplace=True)
prediction.head(10)
```
